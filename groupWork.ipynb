Compensation vs. Years of Experience: This will show how compensation changes with experience for each major role (Software Engineer, Data Scientist, Product Manager).

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Filter the dataset for relevant roles
relevant_roles = ['Software Engineer', 'Data Scientist', 'Product Manager']
filtered_data = data[data['title'].isin(relevant_roles)]

# Set the plot style
sns.set(style="whitegrid")

# Plot: Compensation vs. Years of Experience for each role
plt.figure(figsize=(10, 6))
sns.scatterplot(x='yearsofexperience', y='totalyearlycompensation', hue='title', data=filtered_data, palette="deep")

# Add titles and labels
plt.title('Compensation vs. Years of Experience Across Roles', fontsize=16)
plt.xlabel('Years of Experience', fontsize=12)
plt.ylabel('Total Yearly Compensation ($)', fontsize=12)
plt.legend(title='Role')
plt.tight_layout()

# Show the plot
plt.show()
```
